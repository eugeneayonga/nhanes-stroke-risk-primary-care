### Cycle Concatenation

```python
# Add cycle column to each dataframe and concatenate row-wise
df_nhanes_list = [
    df_nhanes_1999_2000,
    df_nhanes_2001_2002,
    df_nhanes_2003_2004,
    df_nhanes_2005_2006,
    df_nhanes_2007_2008,
    df_nhanes_2009_2010,
    df_nhanes_2011_2012,
    df_nhanes_2013_2014,
    df_nhanes_2015_2016,
    df_nhanes_2017_2018
]

# Period names corresponding to each dataframe
periods = [
    '1999-2000',
    '2001-2002',
    '2003-2004',
    '2005-2006',
    '2007-2008',
    '2009-2010',
    '2011-2012',
    '2013-2014',
    '2015-2016',
    '2017-2018'
]

# Add cycle column to each dataframe and concatenate
dfs_with_cycle = []
for df, period in zip(df_nhanes_list, periods):
    df_copy = df.copy()
    df_copy['CYCLE'] = period
    dfs_with_cycle.append(df_copy)

# Concatenate all dataframes row-wise (stack them vertically)
df_nhanes_1999_2018 = pd.concat(dfs_with_cycle, ignore_index=True)
```
